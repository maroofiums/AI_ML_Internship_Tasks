## **3. Feature Enginerring**

```python
train['chol_trestbps_ratio'] = train['chol'] / train['trestbps']
test['chol_trestbps_ratio'] = test['chol'] / test['trestbps']

train['risk_score'] = train['cp'] + train['exang'] + train['oldpeak'] + train['ca'] + train['thal']
test['risk_score'] = test['cp'] + test['exang'] + test['oldpeak'] + test['ca'] + test['thal']

train['age_thalach_interaction'] = train['age'] * train['thalach']
test['age_thalach_interaction'] = test['age'] * test['thalach']
```

```python
y = train['target']
train_features = train.drop('target', axis=1)

test_ids = test['Id']
test_features = test.drop('Id', axis=1)

categorical_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
numerical_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'chol_trestbps_ratio', 'risk_score', 'age_thalach_interaction']
```

```python
train_features = pd.get_dummies(train_features, columns=categorical_cols, drop_first=True)
test_features = pd.get_dummies(test_features, columns=categorical_cols, drop_first=True)

train_features, test_features = train_features.align(test_features, join='left', axis=1, fill_value=0)
```

```python
scaler = StandardScaler()

train_features[numerical_cols] = scaler.fit_transform(train_features[numerical_cols])
test_features[numerical_cols] = scaler.transform(test_features[numerical_cols])
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_features, y, test_size=0.2, random_state=42)

print("Training data split into training and validation sets.")
print(f"X_train shape: {X_train.shape}")
print(f"X_val shape: {X_val.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_val shape: {y_val.shape}")
```

## **5. Defining Classification Models**

```python
lr_model = LogisticRegression(random_state=42, solver='liblinear') # Added solver for LogisticRegression
knn_model = KNeighborsClassifier()
svc_model = SVC(random_state=42)
dtc_model = DecisionTreeClassifier(random_state=42)
rfc_model = RandomForestClassifier(random_state=42)
gbc_model = GradientBoostingClassifier(random_state=42)
abc_model = AdaBoostClassifier(random_state=42, estimator=DecisionTreeClassifier(max_depth=1)) # Added estimator for AdaBoost
gnb_model = GaussianNB()
```

## **6. Hyperparameter Tuning**


```python

param_grids = {
    'Logistic Regression': {
        'C': [0.01, 0.1, 1, 10, 100],
        'solver': ['liblinear', 'lbfgs']
    },
    'K-Nearest Neighbors': {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance']
    },
    'Support Vector Classifier': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf']
    },
    'Decision Tree': {
        'max_depth': [None, 10, 20, 30],
        'min_samples_leaf': [1, 2, 4]
    },
    'Random Forest': {
        'n_estimators': [50, 100, 150],
        'max_depth': [10, 20, None]
    },
    'Gradient Boosting': {
        'n_estimators': [50, 100, 150],
        'learning_rate': [0.01, 0.1, 0.2]
    },
    'AdaBoost': {
        'n_estimators': [50, 100, 150],
        'learning_rate': [0.01, 0.1, 1]
    },
    'Gaussian Naive Bayes': {}
}

models = {
    'Logistic Regression': lr_model,
    'K-Nearest Neighbors': knn_model,
    'Support Vector Classifier': svc_model,
    'Decision Tree': dtc_model,
    'Random Forest': rfc_model,
    'Gradient Boosting': gbc_model,
    'AdaBoost': abc_model,
    'Gaussian Naive Bayes': gnb_model
}

grid_search_results = {}
best_estimators = {}

print("Starting GridSearchCV for each model...")

for name, model in models.items():
    print(f"\n--- Tuning {name} ---")
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grids[name],
        cv=5,
        scoring='accuracy',
        n_jobs=-1,
        verbose=1
    )

    grid_search.fit(X_train, y_train)

    print(f"Best parameters for {name}: {grid_search.best_params_}")
    print(f"Best cross-validation accuracy for {name}: {grid_search.best_score_:.4f}")

    grid_search_results[name] = grid_search
    best_estimators[name] = grid_search.best_estimator_

print("\nGridSearchCV completed for all models.")
```

## **7. Model Training and Evaluation**

```python

model_performance = {}

print("Evaluating models on the validation set...")

for name, model in best_estimators.items():
    y_pred = model.predict(X_val)
    accuracy = accuracy_score(y_val, y_pred)
    model_performance[name] = accuracy

print("\nModel Performance on Validation Set:")
for name, accuracy in model_performance.items():
    print(f"{name}: {accuracy:.4f}")

best_model_name = max(model_performance, key=model_performance.get)
best_accuracy = model_performance[best_model_name]

print(f"\nThe best performing model(s) on the validation set is/are: {best_model_name} with an accuracy of {best_accuracy:.4f}")
```
